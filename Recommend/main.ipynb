```python
import pandas as pd
```

```python
movies=pd.read_csv('/workspaces/CodSoft/Recommend/maindata.csv')
```

```python
movies.head(10)
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies=movies[['director_name', 'movie_title', 'comb', 'genres']]
```

```python
movies
```

```python
movies['tags'] = movies['comb']+movies['genres']
```

```python
movies
```

```python
new_data  = movies.drop(columns=['comb', 'genres'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=6010, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['movie_title']=="avatar"].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].movie_title)
```

```python
def recommand(movies):
    index=new_data[new_data['movie_title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].movie_title)
```

```python
recommand("men in black")
```
